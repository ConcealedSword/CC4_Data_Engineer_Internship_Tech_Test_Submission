```python
import pandas as pd
restaurants=pd.DataFrame()
# Assuming df is your DataFrame
# Define a function to extract information from the "restaurants" column
def extract_restaurant_info(row):
    # Initialize default values
    restaurant_id = None
    restaurant_name = None
    restaurant_url = None
    restaurant_events = []
    restaurant_country_code = None
    restaurant_city = None
    restaurant_user_rating_votes = None
    restaurant_user_aggregate_rating = None
    restaurant_cuisines = None
    
    # Check if the row is not an empty list
    if isinstance(row, list) and len(row) > 0:
        restaurant_data = row[0]  # Get the first dictionary
        
        # Extract information if available
        if 'restaurant' in restaurant_data:
            restaurant_info = restaurant_data['restaurant']
            restaurant_id = restaurant_info['R']['res_id']
            restaurant_name = restaurant_info['name']
            restaurant_url = restaurant_info['url']
            restaurant_country_code = restaurant_info['location']['country_id']
            restaurant_city = restaurant_info['location']['city']
            restaurant_user_rating_votes = restaurant_info['user_rating']['votes']
            restaurant_user_aggregate_rating = restaurant_info['user_rating']['aggregate_rating']
            restaurant_cuisines = restaurant_info['cuisines']

            
            # Extract events if available, otherwise return an empty list
#             restaurant_events = restaurant_info.get('zomato_events', [])
    
    return pd.Series([restaurant_id, restaurant_name, restaurant_country_code, restaurant_city,restaurant_user_rating_votes,restaurant_user_aggregate_rating, restaurant_cuisines,restaurant_url])

# Apply the function to create new columns
restaurants[['restaurant_id', 'restaurant_name', 'restaurant_country_code','restaurant_city','restaurant_user_rating_votes','restaurant_user_aggregate_rating','restaurant_cuisines','restaurant_url']] = df['restaurants'].apply(extract_restaurant_info)

# Drop the original "restaurants" column if no longer needed
# df.drop('restaurants', axis=1, inplace=True)
```

```python
def extract_start_date(row):
    events = row  # Access the list directly
    if events and events[0].get("restaurant"):
        restaurant_info = events[0]["restaurant"]
        if "zomato_events" in restaurant_info:
            first_event = restaurant_info["zomato_events"][0]  # Get the first event and its start_date
            return first_event["event"]["start_date"]
    return None

# Apply the function to create a new column "start_date"
df["start_date"] = df["restaurants"].apply(extract_start_date)

# Filter rows with non-NaN dates that start with "2019-04"
april_2019_data = df[df["start_date"].notna() & df["start_date"].str.startswith("2019-04")]

# Display the filtered DataFrame
display(df)
display(april_2019_data)
```

```python
def extract_event_info(row):
    # Initialize default values
    event_id=None
    restaurant_id = None
    restaurant_name = None
    restaurant_url = None
    restaurant_events = []
    restaurant_country_code = None
    restaurant_city = None
    restaurant_user_rating_votes = None
    restaurant_user_aggregate_rating = None
    restaurant_cuisines = None
    photo_url=None
    event_title=None
    event_start_date=None
    event_end_date=None
    
    # Check if the row is not an empty list
    if isinstance(row, list) and len(row) > 0:
        restaurant_data = row[0]  # Get the first dictionary
        
        # Extract information if available
        if 'restaurant' in restaurant_data:
            restaurant_info = restaurant_data['restaurant']
            
            event_id=restaurant_info['zomato_events'][0]['event']['event_id']
            photo_url = restaurant_info['zomato_events'][0]['event']['photos'][0]['photo']['url']
#             event_title=restaurant_info['zomato_events'][0]['event']['end_date']
            event_start_date=restaurant_info['zomato_events'][0]['event']['start_date']
            event_end_date=restaurant_info['zomato_events'][0]['event']['end_date']
            restaurant_id = restaurant_info['R']['res_id']
            restaurant_name = restaurant_info['name']

            

    
    return pd.Series([event_end_date,event_start_date,photo_url, event_id,restaurant_id, restaurant_name, ])

```

```python
events=pd.DataFrame()
events[['event_end_date','event_start_date', 'photo_url','event_id','restaurant_id', 'restaurant_name']] = april_2019_data['restaurants'].apply(extract_event_info)
events
```

```python
def extract_rating_info(row):
    # Initialize default values
    aggregate_rating=None
    rating_text=None
    
    # Check if the row is not an empty list
    if isinstance(row, list) and len(row) > 0:
        restaurant_data = row[0]  # Get the first dictionary
        
        # Extract information if available
        if 'restaurant' in restaurant_data:
            restaurant_info = restaurant_data['restaurant']
            aggregate_rating = restaurant_info['user_rating']['aggregate_rating']
            rating_text = restaurant_info['user_rating']['rating_text']

            

    
    return pd.Series([aggregate_rating,rating_text])
```

```python
ratings=pd.DataFrame()
ratings[['aggregate_rating','rating_text']] = df['restaurants'].apply(extract_rating_info)
ratings
```

```python
ratings['aggregate_rating'] = pd.to_numeric(ratings['aggregate_rating'], errors='coerce')

rating_stats = ratings.groupby('rating_text')['aggregate_rating'].agg(['mean', 'min', 'max']).reset_index()

# Rename columns for clarity
rating_stats.columns = ['rating_text', 'mean_rating', 'min_rating', 'max_rating']

# Display the rating statistics
display(rating_stats)
```
